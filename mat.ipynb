```python
x1=np.array([1,2,3,4,5])
x2=np.array([4,6,9,2,5])
plt.plot(x1,x2,marker='o')
```

```python
plt.plot(x1,x2,'*:r') # marker *, line dotted ,color r

#line solid,dotted,dashed
```

```python
plt.plot(x1,x2,'*--b')
```

```python
#marker properties
plt.plot(x1,x2,marker='o',ms=15) #ms=marker size
```

```python
#mfc= marker face color , mec= marker edge color
plt.plot(x1,x2,marker='o',ms=15,mfc='r',mec='y')
```

```python
#line properties #dotted,solid,dashed
plt.plot(x1,x2,marker='o',linestyle='dotted',linewidth=4)
```

```python
plt.plot(x1,x2,marker='o',linestyle='dashed',linewidth=3,color="r")
```

```python
#2 lines in one
x1=[1,2,3,4]
y1=[4,6,7,9]
x2=[3,4,5,7]
y2=[6,7,8,4]
plt.plot(x1,y1,color='r')
plt.plot(x2,y2,color='g')
```

```python
height=[150,156,149,170]
weight=[45,54,67,62]
plt.plot(height,weight,color='g')
plt.xlabel("Height in cm",color='r',fontsize=10)
plt.ylabel("Weight in kg",color='r',fontsize=10)
plt.title("Height-Weight Graph",color='r',fontsize=15)
```

```python
velocity=[40,30,45,55,49]
displacement=[100,105,150,140,110]
plt.plot(velocity,displacement)
plt.xlabel("Velocity in m/s",color='r',fontsize=10)
plt.ylabel("Displacement in m",color='r',fontsize=10)
plt.title("Velocity-Displacement Graph",color='g',fontsize=15)
plt.grid()  # to convert grid
```

```python
velocity=[40,30,45,55,49]
displacement=[100,105,150,140,110]
plt.plot(velocity,displacement)
plt.xlabel("Velocity in m/s",color='r',fontsize=10)
plt.ylabel("Displacement in m",color='r',fontsize=10)
plt.title("Velocity-Displacement Graph",color='g',fontsize=15)
plt.grid(axis='x',linestyle='dotted',linewidth=3,color='r')  # to convert grid
```
