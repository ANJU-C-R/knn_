```python
import numpy as numpy
import pandas as pd 
df=pd.read_csv('company_sales_data.csv')
df
```

```python
#1. Read Total profit of all months and show it using a line plot
import matplotlib.pyplot as plt
x=df['month_number']
y=df['total_profit']
plt.plot(x,y)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.legend(loc='lower right')  #here labels are not given
plt.xticks(x)
plt.yticks([100000,200000,300000,400000,500000,600000])  #values we need in axis
```

```python
#2. Get total profit of all months and show line plot with the following Style properties
#  Generated line plot must include following Style properties:
#  Line Style dotted and Line-color should be red
#  Show legend at the lower right location.
#  X label name = Month Number
#  Y label name = Sold units number
#  Add a circle marker.
#  Line marker color as red
#  Line width should be 3
x=df['month_number']
y=df['total_profit']
plt.plot(x,y,linestyle='dotted',color='r',marker='o',mfc='r',mec='r',linewidth=3,label='Profit Data')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.legend(loc='lower right')
plt.xticks(x)
plt.yticks([100000,200000,300000,400000,500000,600000])
```

```python
#4.Read toothpaste sales data of each month and show it using a scatterplot
#y=df['toothpaste']
plt.scatter(x,df['toothpaste'])
plt.xlabel('Month Number')
```

```python
#5.Read face cream and facewash product sales data and show it using the bar chart with grid line
facecream=df['facecream']
facewash=df['facewash']
plt.bar([a-0.14 for a in x],facecream,width=0.25,align='edge',label='Face Cream Sales Data')
plt.bar([a+0.14 for a in x],facewash,width=0.25,align='edge',label='Face Wash Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Profit')
plt.title('Face Cream and Face Wash')
plt.legend(loc='upper right')
plt.tight_layout()
plt.xticks(x)
plt.grid()
```

```python
#7.Calculate total sale data for last year for each product and show it using a Pie chart
labels=['Face Cream','Face Wash','Bathing Soap','Tooth Paste','Shampoo','Moisturizer']
sales=[df['facecream'].sum(),df['facewash'].sum(),df['bathingsoap'].sum(),df['toothpaste'].sum(),df['shampoo'].sum(),df['moisturizer'].sum()]
plt.pie(sales,labels=labels,autopct='%1.01f%%')
plt.legend(loc='upper right')
plt.tight_layout()
```
