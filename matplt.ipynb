```python
#customer1.txt
import numpy as np
import pandas as pd
df=pd.read_csv('customer1.txt')
df.columns=['id','fname','lname','age','prof','location']
df
```

```python
#each location count  =====> count visualization bar
df1=df['location'].value_counts()
df1
```

```python
import matplotlib.pyplot as plt
plt.bar(df1.index,df1.values)  #(x,y)
plt.xlabel('Locations')
plt.ylabel('Count')
plt.title('Location Count Data')
```

```python
plt.pie(df1.values,labels=df1.index,autopct='%1.0f%%')
plt.legend(loc='lower right')
plt.tight_layout()
```
